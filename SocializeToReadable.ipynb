# Label Data

```python
def tokenizer(text):
  text = text.replace(" ", "")
  words = word_tokenize(text, engine='attacut') #Neural network
  # stop_words = thai_stopwords()
  # filter = []
  # for word in words:
  #   if word not in stop_words:
  #     filter.append(word)
  # return filter  
  return words
```

```python
def tokenizerMM(text):
  text = text.replace(" ", "")
  words = word_tokenize(text, engine='newmm') #Neural network
  # stop_words = thai_stopwords()
  # filter = []
  # for word in words:
  #   if word not in stop_words:
  #     filter.append(word)
  # return filter  
  return words
```

```python
df['tokenizeMM'] = df['sentence'].apply(tokenizerMM)
```

```python
AttaCutDict = {}
for i, line in df.iterrows():
  # attacut = word_tokenize(line['tokenize'], engine="attacut")
  for word in line['tokenize']:
    try:
      AttaCutDict[word] += 1
    except:
      AttaCutDict[word] = 1
```

```python
mm = {}
for i, line in df.iterrows():
  # attacut = word_tokenize(line['tokenize'], engine="attacut")
  for word in line['tokenizeMM']:
    try:
      mm[word] += 1
    except:
      mm[word] = 1
```

```python
mm =  sorted(mm.items(), key=lambda x:x[1], reverse=True)
```
